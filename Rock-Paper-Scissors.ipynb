```python
import random

def get_computer_choice():
    return random.choice(["rock", "paper", "scissors"])

def determine_winner(user, computer):
    if user == computer:
        return "tie"
    
    if (user == "rock" and computer == "scissors") or \
       (user == "scissors" and computer == "paper") or \
       (user == "paper" and computer == "rock"):
        return "user"
    
    return "computer"

def main():
    user_score = 0
    computer_score = 0

    print("\nğŸ® Welcome to Rock-Paper-Scissors Game!")
    print("Instructions:")
    print("Type: rock, paper, scissors\n")

    while True:
        user = input("ğŸ‘‰ Enter your choice (rock/paper/scissors): ").lower()
        
        if user not in ["rock", "paper", "scissors"]:
            print("âŒ Invalid choice! Try again.\n")
            continue

        computer = get_computer_choice()
        print(f"ğŸ§ You chose: {user}")
        print(f"ğŸ’» Computer chose: {computer}")

        result = determine_winner(user, computer)

        if result == "tie":
            print("ğŸ˜ It's a tie!")
        elif result == "user":
            print("ğŸ‰ You win this round!")
            user_score += 1
        else:
            print("ğŸ’€ Computer wins this round!")
            computer_score += 1

        print(f"\nğŸ“Š Scoreboard â†’ You: {user_score} | Computer: {computer_score}\n")

        play_again = input("ğŸ” Do you want to play again? (yes/no): ").lower()
        if play_again != "yes":
            print("\nğŸ Final Score:")
            print(f"â¡ You: {user_score} | Computer: {computer_score}")
            print("ğŸ‰ Thanks for playing! Goodbye!")
            break

main()
```
